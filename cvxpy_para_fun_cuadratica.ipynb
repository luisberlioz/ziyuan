Queremos trabajar con funciones del tipo

 $f(x) = a + b ^T x + \frac12  x^T  P  x $    
 $x \in R^n, P \in S_+^n$

 Recordemos que $\nabla^2 f(x) = P$ y $\nabla f(x) = Px + b$

 Si $\det(P) \neq 0$, entonces tenemos que la unica solucion de $\nabla f(x)$ = 0 es $x = -P^{-1}b$.

```python
n = 3 # dimension del problema
a = 2
b = np.array([1,2,3])
P = np.array([[6,2,3],[2,16,5],[3,5,10]])

# Implementacion de f(x)
def quad(x):
  '''
  a: escalar
  b: vector
  P: matriz positiva definida
  x: vector
  '''
  return a + b.dot(x) + 1/2*x.dot(P.dot(x))

# para evaluar la funcion
print(quad(np.array([0,-1,0])))   # tiene que dar 2

# el minimo de esta funcion esta dado por P.inv * b
print("El minimo ocurre en el punto: ", -np.linalg.inv(P).dot(b))
print("El valor minimo de la funcion es:", quad(-np.linalg.inv(P).dot(b)))
```

```python
# Creamos una variable cvxpy

x = cp.Variable(n)

# Creamos un problema
objective = cp.Minimize(a + b @ x + 0.5 * cp.quad_form(x, cp.psd_wrap(P)))

# el problema sin restricciones
restricciones = []

# definimos el problema
prob = cp.Problem(objective, restricciones)

# resolvemos el problema
result = prob.solve()
print(result)
print(x.value)
```
